```python
def neg_fix(row):
    if row['age'][0] == '-':
        val = row['age'][1:]
    else:
        val = row['age']
    return val
```

```python
intakes['no_negs'] = intakes.apply(neg_fix, axis=1)
```

```python
intakes[intakes['age'] == '-1 years']
```

```python
def norm_age(row):
    # if it ends with year/years, multiply by 12 and enter into month col; enter number into year col
    if (row['no_negs'].split(' ')[1] == 'year') or (row['no_negs'].split(' ')[1] == 'years'):
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val * 12, 2)
        age_years = round(age_val, 2)
    # if it ends with month/months, enter number into month col; divide by 12 and place into year col
    elif (row['no_negs'].split(' ')[1] == 'month') or (row['no_negs'].split(' ')[1] == 'months'):
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val, 2)
        age_years = round(age_val/12, 2)
    # if it ends with week/weeks, divide by four and place into month col; leave default value in year col
    elif (row['no_negs'].split(' ')[1] == 'week') or (row['no_negs'].split(' ')[1] == 'weeks'):
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val/4, 2)
        age_years = 0
    # else (if it ends with day or days), leave default value in both cols
    else:
        age_months = 0
        age_years = 0
    return (age_months, age_years)
```

```python
def norm_age_in(row):
    # if it ends with year/years, multiply by 12 and enter into month col; enter number into year col
    if row['no_negs'].split(' ')[1] in ['year', 'years']:
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val * 12, 2)
        age_years = round(age_val, 2)
    # if it ends with month/months, enter number into month col; divide by 12 and place into year col
    elif row['no_negs'].split(' ')[1] in ['month', 'months']:
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val, 2)
        age_years = round(age_val/12, 2)
    # if it ends with week/weeks, divide by four and place into month col; leave default value in year col
    elif row['no_negs'].split(' ')[1] in ['week', 'weeks']:
        age_val = int(row['no_negs'].split(' ')[0])
        age_months = round(age_val/4, 2)
        age_years = 0
    # else (if it ends with day or days), leave default value in both cols
    else:
        age_months = 0
        age_years = 0
    return (age_months, age_years)
```

```python
intakes[['age_months', 'age_years']] = intakes.apply(norm_age, axis=1, result_type='expand').T.values
```
