```python
# imports
import pandas as pd
from datetime import *
```

```python
# read in intakes
intakes_df = pd.read_csv('../sourceData/intakes.csv')
```

```python
# view data
intakes_df.head()
```

```python
# using 24 hr time instead of 12 hr time
intakes_df['DateTime'] = intakes_df['DateTime'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p').strftime('%m/%d/%Y %H:%M'))
# changing col to datetime
intakes_df['DateTime'] = intakes_df['DateTime'].apply(pd.to_datetime)
# dropping the doubled col
intakes_df = intakes_df.drop(['MonthYear'], axis=1)
intakes_df.head()
```

```python
# checking dtypes
intakes_df.dtypes
```

```python
# checking for nans
intakes_df.isnull().sum()
```

```python
# looking at the null value in the 'Sex upon intake' col
intakes_df[intakes_df['Sex upon Intake'].isnull() == True]
```

```python
# changing the null value in the 'Sex upon intake' col to 'unknown'
intakes_df['Sex upon Intake'] = intakes_df['Sex upon Intake'].fillna(value = 'Unknown')
intakes_df[intakes_df['Sex upon Intake'].isnull() == True]
```

```python
# changing the null values in the 'Name' col to 'unknown'
intakes_df['Name'] = intakes_df['Name'].fillna(value = 'Unknown')
intakes_df[intakes_df['Name'].isnull() == True]
```

```python
# recheck for nans
# checking for nans
intakes_df.isnull().sum()
```

```python
# changing col names
intakes_df.columns = [
    'id', 'name_intake', 'dateTime_intake', 'loc_found', 'intake_type', 'intake_condition', 'animal_type', 
    'sex_intake', 'age', 'breed', 'colour'
]
```

```python
intakes_df.head()
```

```python
# removing the unexplained asterisks
names = [x if x[0] != '*' else x[1:] for x in intakes_df['name_intake']]
intakes_df['name_intake'] = names
intakes_df.head()
```

```python
# saving to processed folder
intakes_df.to_csv('../cleanedSources/intakeCleaned.csv', index = False)
```
